```python
import sys
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

print(f'Payton {sys.version}')
print(f'NumPy: {np.__version__}')
print(f'Matplotlib: {matplotlib.__version__}')
```

```python
# generate 2D meshgrid
nx, ny = (100,100) # how many points in the x & y direction
x=np.linspace(0,10,nx)
y=np.linspace(0,10,ny)

xv, yv = np.meshgrid(x,y)
```

```python
# define a function to plot xy^2
def f(x,y):
    return x*(y**2)
```

```python
# calculate z value for each x,y point
z = f(xv,yv)
```

```python
# make a colour plot to display the data
plt.figure(figsize=(14,12))
plt.pcolor(xv, yv, z)
plt.title('2D Colour Plot of f(x,y)=xy^2')
plt.colorbar()
plt.show()
```

```python
# generate a new 2D meshgrid for the gradient to avoid calculating \
# the gradient for every point of the previous meshgrid
nx, ny = (10,10)
x=np.linspace(0,10,nx)
y=np.linspace(0,10,ny)
xg, yg = np.meshgrid(x,y)
```

```python
# Calculate the gradient of f(x,y)
Gy, Gx = np.gradient(f(xg,yg)) # it returns rows,columns but in the context of a cartesian plane the rows are the y axis and the columns are the x axis
```

```python
# make a colour plot to display the data
plt.figure(figsize=(14,12))
plt.pcolor(xv, yv, z)
plt.title('Gradient of f(x,y)=xy^2')
plt.colorbar()
plt.quiver(xg, yg, Gx, Gy, scale=1000, color='w')
plt.show()
```

```python
# double check - calculate the gradient of f(x,y)=xy^2

def ddx(x,y):
    return y**2

def ddy(x,y):
    return 2*x*y
```

```python
Gx = ddx(xg, yg)
Gy = ddy(xg, yg)
```

```python
# make a colour plot to display the data
plt.figure(figsize=(14,12))
plt.pcolor(xv, yv, z)
plt.title('Plot of [y^2, 2xy]')
plt.colorbar()
plt.quiver(xg, yg, Gx, Gy, scale=1000, color='w')
plt.show()
```
